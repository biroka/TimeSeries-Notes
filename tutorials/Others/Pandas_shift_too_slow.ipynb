# Native numpy implementation
<hr style="border:2px solid black"> </hr>

```python
def shift(xs, n):
    e = np.empty_like(xs)
    
    if n >=0:
        e[:n] = np.nan
        e[n:] = xs[:-n]
    else:
        e[n:] = np.nan
        e[:n] = xs[-n:]
    
    return e
```

```python
dfa = df.values
```

```python
%timeit
shift(dfa, 6)
```
